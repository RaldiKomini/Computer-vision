```python
import os
import cv2
import matplotlib.pyplot as plt
from collections import defaultdict

labels_path = "./cmpe258-galaxies-5/train/labels"
images_path = "./cmpe258-galaxies-5/train/imagess"


class_to_file = {}
for file in os.listdir(labels_path):
    if file.endswith(".txt"):
        with open(os.path.join(labels_path, file)) as f:
            for line in f:
                class_id = int(line.split()[0])
                if class_id not in class_to_file:
                    class_to_file[class_id] = file
                # Stop once we have one image per class
                if len(class_to_file) == len(os.listdir(labels_path)):
                    break

for class_id, label_file in sorted(class_to_file.items()):
    img_file = label_file.replace(".txt", ".jpg")
    img_path = os.path.join(images_path, img_file)
    
    img = cv2.imread(img_path)
    if img is None:
        continue
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img)
    plt.title(f"Class {class_id}")
    plt.axis("off")
    plt.show()
```
